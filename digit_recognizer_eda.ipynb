```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_dig(index):
  image=df_t.iloc[index,1:].values.reshape(28,28)
  label=df_t.iloc[index,0]
  plt.imshow(image,cmap='gray')
  plt.title(f"Label: {label}")
  plt.axis('off')
  plt.show()

for i in range(10):
  plot_dig(i)
```

```python
pix=df_t.iloc[:,1:]
pix.describe().T[['mean','std', 'min', 'max']].head()
```

```python
pix.values.flatten()
```

```python
np.sum(pix.values==0)/pix.size
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
pix_scaled=StandardScaler().fit_transform(pix)
pca=PCA(n_components=2)
components=pca.fit_transform(pix_scaled)

plt.scatter(components[:,0],components[:,1],c=df_t['label'],cmap='tab10', alpha=0.6)
plt.title("PCA- Projection of digits")
plt.colorbar()
plt.show()
```

Each point represents a digit image, colored by its label (0–9).

If you see distinct clusters, it means the digits are linearly separable to some extent — even in just two dimensions.

Overlapping clusters (e.g., 3 vs 5 or 4 vs 9) suggest those digits share similar pixel patterns and may be harder for models to distinguish.

```python
print(pca.explained_variance_ratio_.sum())
```

```python
#PIXEL STATISTICS
df_t['mean_intensity']=df_t.iloc[:,1:].mean(axis=1)
df_t['nonzero_count']=(df_t.iloc[:,1:]>0).sum(axis=1)
```

```python
#SYMMETRY FEATURES
def symmetry_score(row):
  img=row.values.reshape(28,28)
  return np.sum(np.abs(img-np.fliplr(img)))/784
df_t['h_symmetry']=df_t.iloc[:,1:785].apply(symmetry_score,axis=1)
print(df_t['h_symmetry'].describe())
```

```python
from scipy.ndimage import center_of_mass
def center_mass(row):
  img=row.values.reshape(28,28)
  return center_of_mass(img)
df_t[['x_center','y_center']]=df_t.iloc[:,1:785].apply(center_mass,axis=1,result_type='expand')
print(df_t[['x_center','y_center']].describe())
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(df_t['h_symmetry'],bins=50,kde=True,color='green')
plt.title("Distribution of Horizontal Symmetry")
plt.xlabel("Horizontal Symmetry")
plt.ylabel("Frequency")
plt.show()
```

Shows how symmetric the digits are overall.
Lower scores = more symmetric.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='label',y='h_symmetry', data=df_t, palette='Set3')
plt.title("Horizontal Symmetry by Digit")
plt.xlabel("Digit")
plt.ylabel("Symmetry Score")
plt.show()
```

This helps identify which digits tend to be more symmetric (e.g., 0, 8) and which are asymmetric (e.g., 5, 7).

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_t,x='x_center',y='y_center', hue='label',palette='tab10', alpha=0.6 )
plt.title("Center of Mass Distribution by Digit")
plt.xlabel("X Center")
plt.ylabel("Y Center")
plt.legend(title='Digit', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

This reveals how digit placement varies — some digits are centered, others lean left/right or top/bottom.
